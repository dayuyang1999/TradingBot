```python
import numpy as np
import matplotlib.pyplot as plt
```

hengpan = the price is bouncing around a certain value intraday. No big change for a entire day.

```python
leverages = [1, 2, 3]
def hengpan(leverages):
    wave_percents = np.random.rand(50)*0.1
    up_percents = {}
    fig, ax = plt.subplots(figsize=(5, 2.7))
    for leverage in leverages:
        up_percent = []
        for i in wave_percents:
            up_percent.append(1+i*leverage)
            up_percent.append(1-i*leverage)
        up_percents[leverage] = up_percent
        ax.plot(np.cumprod(np.array(up_percent)), label=f"leverage = x {leverage}")
    ax.set_title("A dollar Value During Hengpan")
    ax.legend()
    
```

```python
hengpan(leverages)
```
